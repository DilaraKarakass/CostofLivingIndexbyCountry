```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
df = pd.read_csv("Cost_of_Living_Index_by_Country_2024.csv")
df.head()
```

```python
df.isnull().sum()
#burada boş bir sütun var mı yok mu ona bakmak istedim
```

```python
df.info() #veri tiplerine bakarak ön işleme tabi tutacağım bir şey var mı bakmak istedim.
```

```python
df_sorted = df.sort_values('Cost of Living Plus Rent Index', ascending=False)

# Çubuk grafiği çizme
plt.figure(figsize=(25, 5))
sns.barplot(x='Country', y='Cost of Living Plus Rent Index', data=df_sorted, palette="plasma")
plt.xticks(rotation=90)
plt.title('Cost of Living Plus Rent Index by Country')
plt.xlabel('Country')
plt.ylabel('Cost of Living Plus Rent Index')
plt.show()
#Yaşam maliyeti ve kira indeksi
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

def _plot_series(series, series_name, series_index=0):
    palette = list(sns.palettes.mpl_palette('Dark2'))
    xs = series['Country']
    ys = series['Restaurant Price Index']

    plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(15, 6), layout='constrained')#Grafiğin boyutunu artırdım
df_sorted = df.sort_values('Restaurant Price Index', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Rank')
plt.ylabel('Restaurant Price Index')
plt.xticks(rotation=90, fontsize=8, ha='center')  # Yazı boyutunu küçülttüm ve ortaladım
plt.show()
#Bu graikte ise restorant fiyat indeksinde birinci sırada İsviçre sonuncu sırada ise Bangladesh olduğunu gördüm.
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

def _plot_series(series, series_name, series_index=0):
    palette = list(sns.palettes.mpl_palette('Dark2'))
    xs = series['Country']
    ys = series['Local Purchasing Power Index']

    plt.bar(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(15, 6), layout='constrained')  # Grafiğin boyutunu artırdım
df_sorted = df.sort_values('Local Purchasing Power Index', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Rank')
plt.ylabel('Local Purchasing Power Index ')
plt.xticks(rotation=90, fontsize=8, ha='center')  # Yazı boyutunu küçülttüm ve ortaladım
plt.show()
#bu grafiğe bakınca yerel satın alma gücü olan ülkelerde birinci sırada Lüxemburg, sonuncu sırada ise Cuba olduğunu gördüm.
```
