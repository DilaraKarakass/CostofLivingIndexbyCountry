```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataFrame = pd.read_csv("Cost_of_Living_Index_by_Country_2024.csv")
```

```python
# En yüksek Cost of Living Index'e sahip 10 ülke
top10_col = dataFrame[['Country', 'Cost of Living Index']].sort_values(by='Cost of Living Index', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Cost of Living Index', y='Country', data=top10_col, palette='viridis')
plt.title('Top 10 Countries by Cost of Living Index')
plt.show()
```

```python
# Kira Endeksi ile Yaşam Maliyeti Endeksi arasındaki ilişki
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rent Index', y='Cost of Living Index', data=dataFrame)
plt.title('Relationship between Rent Index and Cost of Living Index')
plt.xlabel('Rent Index')
plt.ylabel('Cost of Living Index')
plt.show()
```

```python
# Yaşam Maliyeti ile Yerel Satın Alma Gücü Endeksi arasındaki ilişki
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Cost of Living Index', y='Local Purchasing Power Index', data=dataFrame)
plt.title('Relationship between Cost of Living Index and Local Purchasing Power Index')
plt.xlabel('Cost of Living Index')
plt.ylabel('Local Purchasing Power Index')
plt.show()
```

```python
# Korelasyon matrisi
plt.figure(figsize=(12, 8))
sns.heatmap(dataFrame.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Different Indices')
plt.show()
```

```python
# En yüksek Restaurant Price Index'e sahip 10 ülke
top10_rpi = dataFrame[['Country', 'Restaurant Price Index']].sort_values(by='Restaurant Price Index', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Restaurant Price Index', y='Country', data=top10_rpi, palette='magma')
plt.title('Top 10 Countries by Restaurant Price Index')
plt.show()
```
